## 1.2. Проверим расчет эффективности обогащения руды
___
Ниже проверка на стадии `Флотации`



<center>Рис.2: Формула расчета эффективности обогащения руды. </center>


___
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

__Напишем функцию для расчета эффективности обогащения__
___
Предусмотрим обработку непредвиденных событий через `try-except` и поставим этому роду событий маркер `-2`, чтобы мы могли далее его анализировать. Аналогичным образом поставим маркер `-1` для тех случаев, когда эффективность выходит за рамки ожидаемых 0-100%. 

```python
def recovery(data):
    try:
        C_rec = data['rougher.output.concentrate_au']
        F_rec = data['rougher.input.feed_au']
        T_rec = data['rougher.output.tail_au']
        recovery_rate = (C_rec * (F_rec - T_rec)) / (F_rec * (C_rec - T_rec))
        if 1 >= recovery_rate >= 0: 
            return(recovery_rate * 100)
        else:
            return(-1)
    except:
        return(-2)
    
```

```python
#аналогичная функция для финальной стадии
def recovery2(data):
    try:
        C_rec = data['final.output.concentrate_au']
        F_rec = data['final.input.feed_au']
        T_rec = data['final.output.tail_au']
        recovery_rate = (C_rec * (F_rec - T_rec)) / (F_rec * (C_rec - T_rec))
        if 1 >= recovery_rate >= 0: 
            return(recovery_rate * 100)
        else:
            return(-1)
    except:
        return(-2)
```

__Вычислим эффективность обогащения на обучающей выборке для признака `rougher.output.recovery`__

```python
data_train['my_recovery'] = data_train.apply(recovery, axis = 1)
# аналогично для финальной стадии
data_train['my_recovery2'] = data_train.apply(recovery, axis = 1)
```

__Заменим значения строк с  NaN в `rougher.output.recovery` на нули__ 

```python
data_train['rougher.output.recovery'] = data_train['rougher.output.recovery'].fillna(0)
```

__Найдем MAE между расчетом и значением признака__
___
MAE совсем незначительна - видимо расчет почти полностью совпадает с фактическими данными.

__Сравним разброс между расчетом и значением признака__
___
Ниже мы можем увидеть, что разница между расчетными и фактическими показателями не высока, вероятнее всего разница только из-за нестандартных данных которым в расчете мы присвоили признаки `-1` и `-2`, оценим эти явления далее. __Оценим есть ли основания полагать, что средние двух совокупностей: расчетной и фактической эффективности не отличаются__
___
Сформулируем статистический тест о равенстве средних:
- H0: Ср.эффективность_расчетная = Ср.эффективность_фактическая
- H1: Ср.эффективность_расчетная != Ср.эффективность_фактическая

Используем "гипотезу о равенстве средних двух генеральных совокупностей" и положим, что вариации совокупностей совпадают.

Определим уровень статистической значимости `alpha = 0.05`

```python
alpha = 0.05
results_hyp_1 = st.ttest_ind(data_train['rougher.output.recovery'], data_train['my_recovery'], equal_var = True)
```

```python
print("p-значение: ", results_hyp_1.pvalue)
if results_hyp_1.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не можем отвергнуть нулевую гипотезу - скорее всего эффективности обогащения руды, расчетная и фактическая, совпадают.")
```

## 3.1. Напишем функцию для вычисления итоговой sMAPE


<center>Рис.7: Формула симметричного среднего абсолютного процентного отклонения.</center>

```python
def smape(predictions, target):
    predictions = pd.Series(predictions)
    target = pd.Series(target)
    total = 0
    for i in range(0, len(target) - 1, 1):
        if (abs(target.values[i]) + abs(predictions.values[i]) ) == 0: #это обход для тех случаев, когда знаменатель равен 0
            total += 0
        else:
            total += abs(target.values[i] - predictions.values[i]) / ( (abs(target.values[i]) + abs(predictions.values[i]) ) / 2)
    return( (total / len(target)) * 100 )
```

### 3.2.3. Проведем кросс-валидацию для модели случайного леса по показателю `sMAPE`.

```python
pods = 3 #измените здесь количество подов для кросс-валидации
pod_size = int(len(data_train_features_rougher) / pods)
```

```python
#для корректной работы необходимо провести переиндексацию
data_train_features_final = data_train_features_final.reset_index(drop=True)
data_train_target = data_train_target.reset_index(drop=True)
```

__На этапе Флотации__

```python
smape_scores_rougher = [''] * pods

for i in range(pods):
    train_features = data_train_features_rougher[(i * pod_size ):(pod_size + pod_size * i)]
    train_target = data_train_target.loc[(i * pod_size ):(pod_size + pod_size * i) - 1, 'rougher.output.recovery'] 
    if i < (pods-1):
        valid_features = data_train_features_rougher[( (i + 1) * pod_size ):(pod_size + pod_size * (i+1))]
        valid_target = data_train_target.loc[((i + 1) * pod_size ):(pod_size + pod_size * (i+1)) - 1, 'rougher.output.recovery'] 
    else:
        valid_features = data_train_features_rougher[:pod_size]
        valid_target = data_train_target.loc[:pod_size, 'rougher.output.recovery']
    
    model_rfr_rough = RandomForestRegressor(max_depth = сlf_forest.best_params_['max_depth'], n_estimators = сlf_forest.best_params_['n_estimators'], random_state = 12356)
    model_rfr_rough.fit(train_features, train_target)
    prediction = model_rfr_rough.predict(valid_features)
    
    smape_scores_rougher[i] = smape(prediction, valid_target)
```

```python
print("Средний показатель sMAPE:")
np.mean(smape_scores_rougher)
```

__На финальном этапе__

```python
smape_scores_final = [''] * pods

for i in range(pods):
    train_features = data_train_features_final[(i * pod_size ):(pod_size + pod_size * i)]
    train_target = data_train_target.loc[(i * pod_size ):(pod_size + pod_size * i) - 1, 'final.output.recovery'] 
    if i < (pods-1):
        valid_features = data_train_features_final[( (i + 1) * pod_size ):(pod_size + pod_size * (i+1))]
        valid_target = data_train_target.loc[((i + 1) * pod_size ):(pod_size + pod_size * (i+1)) - 1 , 'final.output.recovery'] 
    else:
        valid_features = data_train_features_final[:pod_size]
        valid_target = data_train_target.loc[:pod_size, 'final.output.recovery']
    
    model_rfr_final = RandomForestRegressor(max_depth = сlf_forest_final.best_params_['max_depth'], n_estimators = сlf_forest_final.best_params_['n_estimators'], random_state = 12)
    model_rfr_final.fit(train_features, train_target)
    prediction = model_rfr_final.predict(valid_features)
    
    smape_scores_final[i] = smape(prediction, valid_target)
```
